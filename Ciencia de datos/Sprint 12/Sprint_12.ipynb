```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#importacion de dataframes
df_geo_data_0=pd.read_csv('/datasets/geo_data_0.csv')
df_geo_data_1=pd.read_csv('/datasets/geo_data_1.csv')
df_geo_data_2=pd.read_csv('/datasets/geo_data_2.csv')
```

```python
#eliminar la columna "id" pues no se utiilizara
region_0 = df_geo_data_0.drop('id', axis = 1)
region_1 = df_geo_data_1.drop('id', axis = 1)
region_2 = df_geo_data_2.drop('id', axis = 1)
```

```python
# entrenar el modelo con los datos de las 3 regiones a su vez
# que se realiza el calculo de raíz del error cuadrático medio (RMSE)
regiones = [region_0, region_1, region_2]
resultados = []
rmse = []
volumen_medio = []
state = np.random.RandomState(12345)

for i in range(len(regiones)):        
    df = regiones[i]
    
    target = df['product']
    features = df.drop(['product'], axis=1)

    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=88
    )
    # Crear y entrenar modelo
    modelo = LinearRegression()
    modelo.fit(features_train, target_train)

    # Predicciones
    pred = modelo.predict(features_valid)

    # Crear dataframe de resultados
    df_result = pd.DataFrame({
        'predicciones': pred,
        'respuestas_correctas': target_valid.values
    })
    resultados.append(df_result)

    vol = df_result['predicciones'].mean()
    volumen_medio.append(vol)

    mse = mean_squared_error(target_valid, pred)
    rmse = mse**.5

    print(f"Volumen medio de reservas de la región {i} = {round(vol,2)}")
    print(f"RMSE de la región {i} = {round(rmse,2)}")
```

```python
# se establecen los condiciones del negocio
presupuesto = 100_000_000
numero_pozos = 200
unidad_barril = 4_500
unidades_perdidas = 111.1
```

```python
# verificar unidades faltantes
unidades_region=[]
for i in range(len(regiones)):     
    unidades_region.append(unidades_perdidas-volumen_medio[i])
    print(f"Region {i} tiene {round(unidades_region[i],2)} faltantes")
```

```python
# ordenar las mejores 200 predicciones en las regiones 
top_200=[]
for i in range(len(regiones)):
    top_200.append(resultados[i].sort_values(by="predicciones",ascending=False).reset_index())
    top_200[i]=top_200[i][:200]
    print(top_200[i].head())
```

```python
#se obtiene el volumne total a partir de las 200 predicciones realizadas
top_200_volumen=[]

for i in range(len(regiones)):
    suma_region=top_200[i]["predicciones"].sum()
    volumen = suma_region * unidad_barril
    top_200_volumen.append(volumen)
    print(f"la volumen de los 200 mejores pozos de la region {i} es {round(top_200_volumen[i],2)}")

```

```python
#se obtiene la ganancia
ganancia=[]
for i in range(len(regiones)):
    gan=top_200_volumen[i]-presupuesto
    ganancia.append(gan)
    print(f"la ganacia de los 200 mejores pozos de la region {i} es {round(ganancia[i],2)}")
```

```python
# se crea un ciclo for para el calculo de ganancias reales
beneficios=[]
for i in range(len(regiones)):
    beneficios_region=[]
    for j in range(1000):
        muestra=resultados[i].sample(n=500, replace=True)
        mejores_200 = muestra.sort_values(by="predicciones", ascending=False).head(200)
        volumen_muestra = mejores_200['respuestas_correctas'].sum() * unidad_barril
        ganancia_muestra = volumen_muestra - presupuesto
        
        beneficios_region.append(ganancia_muestra) 
    
    beneficios.append(beneficios_region)
```

```python
#se realiza el bootstrapping
for i in range(len(regiones)):
    media=np.mean(beneficios[i])
    cuartil_1 = np.quantile(beneficios[i], 0.025) 
    cuartil_2 = np.quantile(beneficios[i], 0.975) 
    probabilidad_perdida = np.mean(np.array(beneficios[i]) < 0)
    porcentaje_perdida = probabilidad_perdida * 100
    
    print(f"Beneficio Promedio en la Región {i} = {round(media,2)}")
    print(f"Intervalo de Confianza al 95% en la Región {i} = {round(cuartil_1,2)} - {round(cuartil_2,2)}") 
    print(f"Riesgo de Pérdida en la Región {i}: {round(porcentaje_perdida,2)}%") 
```
