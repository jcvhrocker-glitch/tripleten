```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
#se lee el archivo csv
df=pd.read_csv("/datasets/Churn.csv")
```

```python
#se ve la informacion de este
df.info()
```

```python
#se separan las columnas por palabras
mapa_nombres_columnas = {
    'RowNumber':'Row Number',
    'CustomerId': 'Customer Id',
    'CreditScore':'Credit Score',
    'NumOfProducts':'Num Of Products',
    'HasCrCard': 'Has Cr Card',
    'IsActiveMember':'Is Active Member',
    'EstimatedSalary': 'Estimated Salary'
}
df.rename(columns=mapa_nombres_columnas, inplace=True)
df.columns
```

```python
#las columnas se convierten a snake_case
df.columns=df.columns.str.lower().str.replace(' ', '_').str.strip(' ')
df.columns
```

```python
df.head()
```

```python
#se eliminan las columnas categoricas
df = df.drop(['row_number', 'customer_id'], axis=1)
df.info()
```

```python
#se establece el target y los features
target = df['exited']
features = df.drop(['exited'], axis=1)
```

```python
features = features.drop(['surname'], axis=1)
features = pd.get_dummies(features, columns=['geography', 'gender'])
```

```python
#se verifican los valores nulos

print("Valores faltantes por columna:")
print(features.isnull().sum())
```

```python
#se verifican valores duplicados
print("Valores duplicados por columna:")
print(features.duplicated().sum())
```

```python
#se verifica si existen valores excedidos
print(features['tenure'].unique())
```

```python
#se rellenan los valores faltantes con la media
features['tenure'].fillna(features['tenure'].mean(),inplace=True)
```

```python
#se separa el data set en valores para entrenamiento y prueba
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345
)
```

```python
#se verifica el desbalance de clases
features_zeros =features_train[target_train == 0]
features_ones =features_train[target_train == 1]
target_zeros = target_train[target_train==0]
target_ones = target_train[target_train==1]

print(features_zeros.shape,features_ones.shape,target_zeros.shape,target_ones.shape)
```

```python
#se aplica un sobremuestreo
repeat = 4
features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)    
features_upsampled= shuffle(features_upsampled,random_state=12345) 
target_upsampled= shuffle(target_upsampled,random_state=12345)
```

```python
#se aplica un submuestreo
features_downsampled = pd.concat(
        [features_zeros.sample(frac=.25, random_state=12345)]
        + [features_ones]
    )
target_downsampled = pd.concat(
        [target_zeros.sample(frac=.25, random_state=12345)]
        + [target_ones]
    )

features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )
```

Perfecto, técnicas cómo sobre-muestreo o sub-muestreo o la modificación de parámetros en algunos modelos para especificar el desbalance ayuda a los modelos a poder generalizar mejor cada caso y obtener un buen desempeño
</div>

```python
#arbol de decisión
f1=0
for i in range(1,21):
    model_tree = DecisionTreeClassifier(max_depth=i,random_state=12345)
    model_tree.fit(features_train, target_train)
    predicted_valid_tree = model_tree.predict(features_valid)
    if  f1_score(target_valid, predicted_valid_tree)>f1:
        f1=f1_score(target_valid, predicted_valid_tree)
        depth=i

model_tree = DecisionTreeClassifier(max_depth=depth,random_state=12345)
model_tree.fit(features_train, target_train)
predicted_valid_tree = model_tree.predict(features_valid)
print(depth,f1_score(target_valid,predicted_valid_tree)) 
```
